```python
from bs4 import BeautifulSoup
import requests
import wget

#paste the url of the page 
for page in range(1,100):
    url = "https://www.dreamstime.com/photos-images/barefoot.html?pg=" + str(page)

    #downloading the page 
    page = requests.get(url)    
    data = page.text
    soup = BeautifulSoup(data)

    #empty list for collection the images path
    list_of_images = []
    for link in soup.find_all('img'):
        if str(link.get('src')):
            list_of_images.append(str(link.get('src')))

    print("Downloading Images")

    #downloading images 
    for i in list_of_images:
        try:
            wget.download(i)
        except:
            pass

    print("Download Done")
```

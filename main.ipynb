```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
```

```python
DATADIR = 'C:/Users/yadav/Desktop/assignment/dataset'
CATEGORIES = ['barefoot', 'not visible', 'shoes']

for category in CATEGORIES:
    path = os.path.join(DATADIR, category)      ## path to cat or dog folder
    for img in os.listdir(path):
        img_arr = cv2.imread(os.path.join(path,img), 0)
        plt.imshow(img_arr)
        break
    break
```

```python
IMG_SIZE = 500

new_arr = cv2.resize(img_arr,(IMG_SIZE,IMG_SIZE))
plt.imshow(new_arr, cmap='gray')
plt.show()
```

```python
training_data = []

def create_training_data():
    for category in CATEGORIES:
        path = os.path.join(DATADIR, category)  
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                img_arr = cv2.imread(os.path.join(path,img), 0)
                new_arr = cv2.resize(img_arr,(IMG_SIZE,IMG_SIZE))
                training_data.append([new_arr, class_num])
            except Exception as e:
                pass
create_training_data()
```

```python
import random
random.shuffle(training_data)
```

```python
cap = cv2.VideoCapture(0)
while True:
    
    ret, frame = cap.read()
      
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
    img = np.array(gray)   
    img = cv2.resize(img, (500, 500))
    img = img.reshape(-1, 500, 500, 1)
    res = model.predict(img)
    index = np.argmax(res)
    output = CATEGORIES[index]
    cv2.putText(gray, output, (30,80),cv2.FONT_HERSHEY_SIMPLEX,2,(209, 80, 0, 255),2)
    cv2.imshow('frame', gray)

    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```
